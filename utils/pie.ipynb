```python
import pie

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

import plotly.express as px
```

```python
scores = pie.get_scores(pie.IMAGE_DIR)
```

```python
# get the index of the zeros and non-zeros
indices_of_zeros  = np.where(scores == 0) 
indices_of_non_zeros = np.where(scores != 0)

indices_of_zeros
```

```python
non_zeros = scores[indices_of_non_zeros]
non_zeros.shape
```

```python
non_zeros.mean()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x = non_zeros, palette = 'rocket')
# plt.title("Box plot of scores in the test set ")
plt.show()
```

```python
maximum = np.max(scores)
print("Max score: ", maximum)
```

```python
idx_maximum = np.where(scores == maximum)[0].item()
idx_maximum
```

```python
import os
```

```python
file_names = os.listdir(pie.IMAGE_DIR)
best_file_name = file_names[idx_maximum]
best_file_path = pie.IMAGE_DIR + best_file_name
best_file_path
```

```python
import cv2
```

```python
best_img = cv2.imread(best_file_path)
plt.imshow(best_img)
plt.axis("off")
```

```python
masks = pie.get_masks(best_file_path)
bboxs = pie.get_bboxs(masks)
percents = pie.get_percent(bboxs)
percents
```

```python
plt.figure(figsize=(10, 5)) 
plt.imshow(best_img)
plt.title(f"Best image (score = {maximum}) \nPredicted percent: " + str(percents))
plt.axis('off')
plt.show()
```

```python
minimum = np.min(non_zeros)
print("Min score: ", minimum)
```

```python
# get the index of the minimum
idx_minimum = np.where(scores == minimum)[0].item()
idx_minimum
```

```python
# get the file name of the image with the minimum score
worst_file_name = file_names[idx_minimum]
worst_file_path = pie.IMAGE_DIR + worst_file_name
```

```python
# read the image
worst_img = cv2.imread(worst_file_path)

plt.imshow(worst_img)
plt.axis("off")
```

```python
masks = pie.get_masks(worst_file_path)
bboxs = pie.get_bboxs(masks)
percents = pie.get_percent(bboxs)
percents
```

```python
gt_bbox_count = []
pred_bbox_count = []

i = 0

for file_name in file_names:
    file_path = pie.IMAGE_DIR + file_name
    masks = pie.get_masks(file_path)
    gt_keypoints = pie.get_gt_keypoints(file_path)

    
    gt_bbox_count.append(gt_keypoints.shape[0])
    pred_bbox_count.append(masks.shape[0])

    i += 1
    
    if i % 100 == 0:
        print(f"Processed {i}/1000 images") 
    
```

```python
gt_bbox_count
```

```python
pred_bbox_count
```

```python
df = pd.DataFrame({'gt_bbox_count': gt_bbox_count, 'pred_bbox_count': pred_bbox_count})
df
```

```python
# scatter plot w hover data
fig = px.scatter(df, 
                 x="gt_bbox_count", 
                 y="pred_bbox_count", 
                 labels={"gt_bbox_count": "Ground truth bbox count", 
                         "pred_bbox_count": "Predicted bbox count"},
                 title="Scatter plot of ground truth bbox count vs predicted bbox count")

# ideal case (y = x)
fig.add_shape(type="line", 
              x0=df["gt_bbox_count"].min(), 
              y0=df["gt_bbox_count"].min(), 
              x1=df["gt_bbox_count"].max(), 
              y1=df["gt_bbox_count"].max(),
              line=dict(color="red", width=2))

fig.update_layout(width=1500, height=800)  
fig.show()
```
