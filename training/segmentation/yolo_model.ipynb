## Load libraries and Configs

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt 
import cv2
```

```python
model = YOLO("./yolov8n-seg.pt")
```

```python
model.train(
    data = "./segmentation.yaml",
    epochs = 20, 
    batch = 5,
    imgsz = 640,
    workers = 2
)
```

```python
# save model
model.save("yolov8n-seg.pt")
```
